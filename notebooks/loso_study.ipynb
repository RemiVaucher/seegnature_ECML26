```python
import glob
import pandas as pd
from seegnature.evaluation import evaluate_model
from seegnature.models import SigLR
from seegnature.models.concurrents import CSPLDA, CSPSVM, ATCNetClassifier, EEGNetClassifier, EEGConformerClassifier
from seegnature.utils import subject_dict_from_files, loso_dataset
```

```python
SEED = 42
SHUFFLING = True
```

#Importing Dataset

Full result by subjects are available in "Summary"

```python
subject_dict = {subject: (X[...,-375:], (y == "encoding_memoranda").astype(int)) for subject, (X, y) in subject_dict_from_files(glob.glob("seegnature_ECML26/data/mne/*")).items()}
```

```python
X, y, splits = loso_dataset(subject_dict, shuffle=SHUFFLING, random_state=SEED)
```

# Sig2LR + $\ell_2$

```python
sig2lr_df = evaluate_model(SigLR(2, time_aug=True, lead_lag_aug=True, random_state=SEED), X, y, cv=splits)
sig2lr_df
```

# Sig2LR + $\ell_1$

```python
sig2lr_lasso_df = evaluate_model(SigLR(2, time_aug=True, lead_lag_aug=True, l1_ratio=1.0, random_state=SEED), X, y, cv=splits)
sig2lr_lasso_df
```

#SIG3LR + $\ell_2$

```python
sig3lr_df = evaluate_model(SigLR(3, time_aug=True, lead_lag_aug=True, random_state=SEED), X, y, cv=splits)
sig3lr_df
```

#Sig3LR + $\ell_1$

```python
sig3lr_lasso_df = evaluate_model(SigLR(3, time_aug=True, lead_lag_aug=True, l1_ratio=1.0, random_state=SEED), X, y, cv=splits)
sig3lr_lasso_df
```

# CSPLDA

```python
csplda_df = evaluate_model(CSPLDA(), X, y, cv=splits)
csplda_df
```

# CSPSVM

```python
cspsvm_df = evaluate_model(CSPSVM(random_state=SEED), X, y, cv=splits)
cspsvm_df
```

# ATCNET

```python
atcnet_df = evaluate_model(ATCNetClassifier(random_state=SEED), X, y, cv=splits)
atcnet_df
```

# EEGNet

```python
eegnet_df = evaluate_model(EEGNetClassifier(random_state=SEED), X, y, cv=splits)
eegnet_df
```

# EEG Conformer

```python
eeg_conformer_df = evaluate_model(EEGConformerClassifier(random_state=SEED), X, y, cv=splits)
eeg_conformer_df
```

# Summary

```python
total_df = pd.concat({
    "sig2lr": sig2lr_df,
    "sig2lr_lasso": sig2lr_lasso_df,
    "sig3lr": sig3lr_df,
    "sig3lr_lasso": sig3lr_lasso_df,
    "csplda": csplda_df,
    "cspsvm": cspsvm_df,
    "atcnet": atcnet_df,
    "eegnet": eegnet_df,
    "eegconformer": eeg_conformer_df,
})
#total_df = total_df.swaplevel(1,0).sort_index()
total_df.index = total_df.index.map(lambda x : (x[0], list(subject_dict.keys())[x[1]]))
total_df
```
