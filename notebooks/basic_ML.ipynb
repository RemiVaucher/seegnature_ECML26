# Data loading

In this part we will load the data. We have multiple participants with a EEG recording each. We will load the data following a mapping format between the participant and a tuple with the EEG recording and the label to be predicted by a classifier.

```python

subject_dict = subject_dict_from_files(SUBJECT_FILES)
print(f"Loaded subjects are {", ".join(subject_dict.keys())}")
```

The data follows a "subject : (EEG, label)" format. The EEG itself is an array of shape (batch, channel, time). The label is an array of strings, here either "control_memoranda" or "encoding_memoranda".

The whole epoch is present, we only want the mental imagery part of it which starts at index -375 on the time axis. Similarly, to have access to additional scores for the models such as the F1-score or the AUC, we should get rid of the strings and transform them into either 0 or 1. This allows us to have a binary classification problem.

```python
subject_dict = {subject: (X[..., -375:], (y == "encoding_memoranda").astype(int)) for subject, (X, y) in subject_dict.items()}
```

```python
import numpy as np
from scipy.signal import welch
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_validate, StratifiedKFold
from typing import Literal


def compute_psd_features(X: np.ndarray, sfreq: float, fmin: float = 1.0, fmax: float = 40.0) -> np.ndarray:

    batch, n_elec, n_times = X.shape

    # Calcul du PSD sur un trial pour connaître la taille de sortie
    freqs, psd_sample = welch(X[0, 0], fs=sfreq, nperseg=min(256, n_times))
    freq_mask = (freqs >= fmin) & (freqs <= fmax)
    n_freqs = freq_mask.sum()

    features = np.zeros((batch, n_elec * n_freqs))

    for i in range(batch):
        trial_psds = []
        for e in range(n_elec):
            _, psd = welch(X[i, e], fs=sfreq, nperseg=min(256, n_times))
            trial_psds.append(psd[freq_mask])
        features[i] = np.concatenate(trial_psds)

    return features


def compute_band_features(X: np.ndarray, sfreq: float) -> np.ndarray:

    BANDS = {
        "delta": (0.5, 4.0),
        "theta": (4.0, 8.0),
        "alpha": (8.0, 13.0),
        "beta":  (13.0, 30.0),
        "gamma": (30.0, 45.0),
    }

    batch, n_elec, n_times = X.shape
    n_bands = len(BANDS)
    features = np.zeros((batch, n_elec * n_bands))

    for i in range(batch):
        trial_feats = []
        for e in range(n_elec):
            freqs, psd = welch(X[i, e], fs=sfreq, nperseg=min(256, n_times))
            band_powers = []
            for (fmin, fmax) in BANDS.values():
                mask = (freqs >= fmin) & (freqs <= fmax)
                band_powers.append(psd[mask].mean() if mask.any() else 0.0)
            trial_feats.append(band_powers)
        features[i] = np.array(trial_feats).flatten()

    return features


def build_lr_pipeline() -> Pipeline:
    return Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(max_iter=1000, solver="lbfgs", C=1.0)),
    ])


def classify_eeg(
    X: np.ndarray,
    y: np.ndarray,
    sfreq: float,
    method: Literal["psd", "bands", "both"] = "both",
    cv: int = 5,
) -> dict:

    assert X.ndim == 3 and X.shape[1] == 62,
    assert len(y) == X.shape[0],

    results = {}

    scoring = ["accuracy", "d2_brier_score", "f1", "roc_auc"]

    if method in ("psd", "both"):
        print("Extraction features PSD...")
        X_psd = compute_psd_features(X, sfreq)
        pipe_psd = build_lr_pipeline()
        cv_strategy = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)
        scores_psd = cross_validate(pipe_psd, X_psd, y, cv=cv_strategy, scoring=scoring,return_train_score = False)
        pipe_psd.fit(X_psd, y)  # fit final sur tout le dataset
        results["psd"] = {
            "pipeline": pipe_psd,
            "cv_scores": scores_psd,
        }
        print(f"[LR + PSD]   Scores: {scores_psd}")

    if method in ("bands", "both"):
        print("Extraction features by frequence band...")
        X_bands = compute_band_features(X, sfreq)
        pipe_bands = build_lr_pipeline()
        cv_strategy = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)
        scores_bands = cross_validate(pipe_bands, X_bands, y, cv=cv_strategy, scoring=scoring,return_train_score = False)
        pipe_bands.fit(X_bands, y)  # fit final sur tout le dataset
        results["bands"] = {
            "pipeline": pipe_bands,
            "cv_scores": scores_bands,
        }
        print(f"[LR + Bandes] Scores: {scores_psd}")

    return results


def predict_eeg(
    X_new: np.ndarray,
    results: dict,
    sfreq: float,
    method: Literal["psd", "bands"] = "psd",
) -> np.ndarray:

    if method == "psd":
        X_feat = compute_psd_features(X_new, sfreq)
    else:
        X_feat = compute_band_features(X_new, sfreq)

    return results[method]["pipeline"].predict(X_feat)
```

```python
np.random.seed(SEED)
sfreq = 128.0  # Hz

results = {}

for subject, (X, y) in subject_dict.items():
    print(f"Treating {subject}")
    results[subject] = classify_eeg(X, y, sfreq=sfreq, method="both", cv=5)
```
